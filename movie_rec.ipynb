# Movies Recommendation System

Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing dataset

```python
movies_df = pd.read_csv("E:/Clg Documents/ML_Coursera/movies_dataset1/imdb_top_1000.csv")
```

Selecting only relevant columns

```python
movies_df1 = movies_df[['Series_Title','Released_Year','Genre','IMDB_Rating']]
```

```python
movies_df1.head()
```

Cheking for null values

```python
movies_df1.isnull().sum()
```

Putting genres in a list

```python
movies_df1['Genre'] = movies_df1.Genre.str.split(',')
movies_df1.head()
```

Copying dataset

```python
Genres_df = movies_df1.copy()
```

For every row, splitting up genres in different columns and inserting 1 for the genres existing for that row.

```python
for index, row in movies_df1.iterrows():
    for genre in row['Genre']:
        Genres_df.at[index, genre] = 1
```

Adding 0 in the remaining genres columns 

```python
Genres_df = Genres_df.fillna(0)
```

```python
Genres_df.head()
```

Taking input of movies and rating from the user

```python
userIp = [{'title':'Interstellar','rating':9},{'title':'Joker','rating':8.2},{'title':'Dangal','rating':8.5},{'title':'Avengers: Endgame','rating':9.5},{'title':'3 Idiots','rating':9.7}]
Ipmovies = pd.DataFrame(userIp)
Ipmovies
```

Selecting the rows from genres dataset which has movies from user input.

```python
userMovies = Genres_df[Genres_df['Series_Title'].isin(Ipmovies['title'].tolist())]
userMovies
```

Resetting the index and keeping only genres columns

```python
userMovies = userMovies.reset_index(drop=True)
userMovies = userMovies.drop(['Released_Year','Genre','Series_Title','IMDB_Rating'], axis=1)
```

```python
userMovies
```

```python
Ipmovies['rating']
```

```python
userProfile = userMovies.transpose().dot(Ipmovies['rating'])
userProfile
```

```python
GenreTable = Genres_df.drop(['Released_Year','Genre','Series_Title','IMDB_Rating'],axis=1)
GenreTable
```

With the input's profile and the complete list of movies and their genres in hand, we're going to take the weighted average of every movie based on the input profile.

```python
recommendationTable_df = ((GenreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df
```

Creating a dataset of all movies and adding a column with values got from above recommendation_df.

```python
rec_movies = movies_df1.copy()
rec_movies['values'] = recommendationTable_df
```

Presenting the recommended movies dataset in the descending order based on values.

```python
rec_movies = rec_movies.sort_values('values',ascending=False)
```
